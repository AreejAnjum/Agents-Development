```python
import os
from langchain.chat_models import init_chat_model
os.getenv("google_genai:gemini-2.5-flash")
model=init_chat_model("google_genai:gemini-2.5-flash")
response=model.invoke("What is the capital of Germany?")
response
```

```python
from langchain.tools import tool
@tool
def get_capital_of_country(city: str) -> str:
    """Get the capital of a country.""" #context for the tool
   
    return  f"It is {city}"
#bind tools to the llms
model_with_tools = model.bind_tools([get_capital_of_country])
```

```python
response=model_with_tools.invoke("What is the capital of Germany?")
print(response)
for tool_call in response.tool_calls:
    #view the tool call details made by the model
    print(f"Tool: {tool_call['name']}") 
    print(f"Targs: {tool_call['args']}")
```

### Tool Execution Loops

```python
#just for understanding, agent does this automatically, but we can also run the tool manually
from langchain.tools import tool

# 1. Create a tool
@tool
def get_weather(city: str) -> str:
    """Get the weather for a given city."""
    return f"The weather in {city} is 72°F and sunny."


# 2. Give/bind the tool to the model
model_with_tools = model.bind_tools([get_weather])


# 3. User asks a question
messages = [
    {"role": "user", "content": "What's the weather in Boston?"}
]


# 4. Model decides which tool to call
ai_msg = model_with_tools.invoke(messages)

# Save model message
messages.append(ai_msg)


# 5. Run the tool manually
for tool_call in ai_msg.tool_calls:
    tool_result = get_weather.invoke(tool_call)
    messages.append(tool_result)


# 6. Send tool result back to model
final_response = model_with_tools.invoke(messages)

print(final_response.text())
```
