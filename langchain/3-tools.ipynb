### Tools
omething the AI agent can use to do work outside its own brain.

The model can think and write, but a tool helps it take action.
omething the AI agent can use to do work outside its own brain.

Calculator → for math
Google Search → for latest information
Database tool → to get saved data
Email tool → to send/read emails
Calendar tool → to check meetings
Python tool → to run code



```python
import os
from langchain.chat_models import init_chat_model
os.getenv("google_genai:gemini-2.5-flash")
model=init_chat_model("google_genai:gemini-2.5-flash")
response=model.invoke("What is the capital of Germany?")
response
```

```python
from langchain.tools import tool
@tool
def get_capital_of_country(city: str) -> str:
    """Get the capital of a country.""" #context for the tool
   
    return  f"It is {city}"
#bind tools to the llms
model_with_tool = model.bind_tools([get_capital_of_country])
```
