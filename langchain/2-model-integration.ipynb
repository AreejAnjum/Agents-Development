### Model integration with OpenAi, Google Gemini and Groq

```python
import os 
from dotenv import load_dotenv
load_dotenv()

os.getenv("GOOGLE_API_KEY")
os.getenv("GROQ_API_KEY")
os.getenv("ANTHROPIC_API_KEY")
```

```python
from langchain.chat_models import init_chat_model
model = init_chat_model("claude-haiku-4-5")
model
```

```python
#   invoke the model
response= model.invoke("Hello How are you")
#response
response.content
```

### Google Gemini Model Integration


```python
from langchain.chat_models import init_chat_model
model = init_chat_model("google_genai:gemini-2.5-flash")
model 
```

```python
response= model.invoke("Hello What are the agents in agentic ai")
response.content
```

### Alternative method of model generation and invoke

```python
from langchain_anthropic import ChatAnthropic

model = ChatAnthropic(model="claude-haiku-4-5")
response = model.invoke("Hello What are the agents in agentic ai")
response.content
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI
model = ChatGoogleGenerativeAI(model="gemini-2.5-flash")
response = model.invoke("Hello What are the agents in agentic ai")
response.content
```
