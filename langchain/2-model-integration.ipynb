## Streaming

Streaming in Agentic AI means:

The AI does not wait until the full answer is ready.

It sends the answer piece by piece, like typing live.

Example without streaming:

User: Explain LangGraph
AI waits...
AI sends full answer after 10 seconds

Example with streaming:

User: Explain LangGraph
AI starts immediately:
"LangGraph is..."
"used for..."
"building agents..."
So streaming is useful because:

User sees response faster
App feels more real-time
Good for chatbots
Good for long answers
Good for showing agent steps/tool progress

In agentic AI, streaming can show:

Thinking step started
Tool is being called
Search result received
Final answer is generating


```python
#time taking comparative to streaming mode
model.invoke("Write me a 10 words paragraph on Artificial Intelligence")
```

```python
#it returns a object response 
model.stream("Write me a 10 words paragraph on Artificial Intelligence")
```

```python
# response in text mode 
# flush=True
#Show the output immediately.This is useful in streaming because you want each chunk to appear as soon as the model sends it.
for chunk in model.stream("Write me a 10 words paragraph on Artificial Intelligence"):
    print(chunk.text, end="|", flush=True)
```

### Batch
Giving the agent many tasks or many inputs at once, instead of one by one. They will process parallel, reduce cost and improve performance

```python
response =model.batch(["Write me a 10 words paragraph on Artificial Intelligence",
                       "What AI stands for",
                       "What is agentic AI"])
for response in response:
    print(response)
```

```python
model.batch(["Write me a 10 words paragraph on Artificial Intelligence",
                       "What AI stands for",
                       "What is agentic AI"],
                    config={'max_concurrency': 2} # limit to 2 parallel calls 
                       )
```
